```python
import numpy as np
import pandas as pd
```

```python
k2020 = pd.read_excel('konto2020.xlsx')
k2021 = pd.read_excel('konto2021.xlsx')
k2022 = pd.read_excel('konto2022.xlsx')
k2023 = pd.read_excel('konto2023.xlsx')
```

Rydde i kolonnene

```python
def drop_columns(dataframe, columns_to_drop):
    columns_to_drop_existing = [col for col in columns_to_drop if col in dataframe.columns]
    return dataframe.drop(columns=columns_to_drop_existing, axis=1)

columns_to_drop = [
    'Unnamed: 3',
    'Unnamed: 4',
    'Unnamed: 5',
    'Inn',
    'Ut',
    'Unnamed: 6'
]
```

```python
k2020 = drop_columns(k2020, columns_to_drop)
k2021 = drop_columns(k2021, columns_to_drop)
k2022 = drop_columns(k2022, columns_to_drop)
k2023 = drop_columns(k2023, columns_to_drop)
```
