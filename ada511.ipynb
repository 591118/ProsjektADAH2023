## Expected values

```python
def expected_value(values, probs):
    values = np.asarray(values)
    weights = np.asarray(probs)
    return (values * probs).sum() / probs.sum()
```

```python
print (expected_value(bankU, prob))
```

```python
print (expected_value(personU, prob))
```

The value you calculated, 20, represents the expected value (or expected utility) of the utility matrix given the corresponding probability matrix. In the context of utility theory or decision theory, the expected value is a measure that represents the average utility of a set of outcomes, weighted by their probabilities.

In your case:

    The utility matrix (values) represents the utility associated with each possible outcome or state.
    The probability matrix (probs) represents the likelihood or probability of each corresponding outcome or state.

The expected value is calculated by multiplying each utility value by its probability, summing up these products, and then dividing by the sum of probabilities.
