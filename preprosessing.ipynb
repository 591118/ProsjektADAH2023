```python
import pandas as pd
```

```python
file_paths = ['konto2020.xlsx' , 'konto2021.xlsx', 'konto2022.xlsx', 'konto2023.xlsx']
```

```python
def merge_excel_files(file_paths, output_file):
    data_frames = []
    
    # Read each Excel file and append its data to the list
    for file_path in file_paths:
        df = pd.read_excel(file_path)
        data_frames.append(df)
    
    # Concatenate all the data frames into a single data frame
    combined_data = pd.concat(data_frames, ignore_index=True)
    
    # Write the combined data to a new Excel file
    combined_data.to_excel(output_file, index=False)
    print(f"Combined data has been saved to {output_file}")

# Output file where the combined data will be saved
output_file = 'alle.xlsx'

merge_excel_files(file_paths, output_file)
```

```python
data = pd.read_excel('alle.xlsx')

# Sort the DataFrame by the 'Date' column
data_sorted = data.sort_values(by='Dato')

# Save the sorted data back to the Excel file
data_sorted.to_excel('alle_sorted.xlsx', index=False)
print("Data sorted by date and saved to alle_sorted.xlsx")
```

```python
rente = pd.read_csv('datasett_KPI_Styringsrente\styringsrenten.csv', delimiter=";", decimal=",")
inflasjon1 = pd.read_csv('datasett_KPI_Styringsrente\kpi_1991_2023.csv', delimiter=";", decimal=",")
inflasjon2 = pd.read_csv('datasett_KPI_Styringsrente\KPI_månedlig_endring.csv', delimiter=";", decimal=",")
```

```python
rente = rente[['TIME_PERIOD', 'OBS_VALUE']]
rente = rente.rename(columns={'TIME_PERIOD': 'Dato', 'OBS_VALUE': 'Rente'})
rente = rente.iloc[1:]

rente
```

```python
# Convert date format in inflasjon1 and inflasjon2 to 'year-month'
inflasjon1['Dato'] = pd.to_datetime(inflasjon1['Dato'], format='%YM%m').dt.strftime('%Y-%m')
inflasjon2['Dato'] = pd.to_datetime(inflasjon2['Dato'], format='%YM%m').dt.strftime('%Y-%m')

# Convert date format in rente to 'year-month'
rente['Dato'] = pd.to_datetime(rente['Dato'], format='%Y-%m-%d').dt.strftime('%Y-%m')
```

```python
# Merge inflasjon1 and inflasjon2 on 'Dato'
merged1 = pd.merge(inflasjon1, inflasjon2, on='Dato', suffixes=('', '_inflasjon2'))

# Merge the result with rente on 'Dato'
final_df = pd.merge(merged1, rente, on='Dato', suffixes=('', '_rente'))
```
